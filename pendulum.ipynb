```python
import torch
import torch.nn as nn
from torch.distributions.categorical import Categorical
from torch.optim import Adam
import numpy as np
import gymnasium as gym
import time as t
import matplotlib.pyplot as plt
import Policy_grad as pen
torch.manual_seed(711)
```

```python
model,out = pen.train(epochs=100,lr=0.005)
```

```python
model.eval()
env_name = 'CartPole-v1'
env = gym.make(env_name,render_mode = "human")
obs,_ = env.reset()
done = False
while True:
    
    # act in the environment
    # print(obs)
    act = torch.argmax(model(torch.as_tensor(obs, dtype=torch.float32))).item()
    # print(act)
    obs, rew, done, terminated, truncated, info= env.step(act)
    if terminated or truncated:
        observation, info = env.reset()
        env.close()
```
